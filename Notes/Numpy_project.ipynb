# Image manipulation using Numpy

```python
import numpy as np
```

```python
from skimage import data
```

```python
data1 = data.cat()
```

```python
data1.shape
```

```python
data1.size
```

```python
data1
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(data1)
```

```python
sp_data1 = data1[100:, 50:, :]
```

```python
plt.imshow(sp_data1)
```

```python
sp1,sp2 = np.split(data1,2)
```

```python
plt.imshow(sp1)
```

```python
plt.imshow(sp2)
```

```python
vsplit = data1[:,:-1,:]
sp3,sp4 = np.split(vsplit,2,axis=1)
```

```python
plt.imshow(sp3)
```

```python
plt.imshow(sp4)
```

```python
plt.imshow(np.concatenate((sp1,sp2)))
plt.show()
```

```python
plt.imshow(np.concatenate((sp3,sp4),axis=1))
plt.show()
```
